```python
import pandas as pd

# 1. Load the data (Ensure your CSV is named exactly this in your data/raw folder)
file_path = '../data/raw/german_credit_data.csv'
df = pd.read_csv(file_path)

# 2. Verify the size (Looking for ~1000 rows, 20+ columns)
print("--- DATASET SHAPE ---")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}\n")

# 3. Verify the features match your proposal
print("--- COLUMN NAMES ---")
print(df.columns.tolist())
print("\n")

# 4. Preview the actual data
display(df.head())
```

```python
# 1. Check the exact text values before we change them
print("Original Target Values:")
print(df['Risk'].value_counts())

# 2. Convert 'Risk' to 1 (Bad/Default) and 0 (Good/Paid)
# (We check for both uppercase and lowercase just to be safe)
df['Risk'] = df['Risk'].replace({'good': 0, 'bad': 1, 'Good': 0, 'Bad': 1})

# 3. Check the new class imbalance percentage
print("\nNew Class Imbalance (1 = Bad, 0 = Good):")
print(df['Risk'].value_counts(normalize=True) * 100)
```

```python
# 1. Apply One-Hot Encoding to all categorical (text) columns
# drop_first=True prevents "dummy variable trap" (multicollinearity)
df_processed = pd.get_dummies(df, drop_first=True)

# 2. Look at how our dataset has transformed
print("--- FEATURE ENGINEERING COMPLETE ---")
print(f"Old shape: {df.shape}")
print(f"New shape (after One-Hot Encoding): {df_processed.shape}\n")

# 3. Preview the new numerical columns
display(df_processed.head())

# 4. Save the finalized artifact to our Git-ignored processed folder
output_path = '../data/processed/german_credit_ml_ready.csv'
df_processed.to_csv(output_path, index=False)
print(f"Success! Cleaned dataset saved to: {output_path}")
```
